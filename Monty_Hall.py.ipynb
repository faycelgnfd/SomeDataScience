```python
from random import *
import matplotlib.pyplot as plt

seed()
def play_game(strategie):
    #1==garder, 2==changer
    porte=[0,1,2]
    
    premier_choix = randint(0,2)
    
    bonne_porte = randint(0,2)
    
    porte.remove(premier_choix)
    
    #limination de porte
    if premier_choix==bonne_porte:
        #elimine une des portes restante au hasard
        porte.remove(porte[randint(0,1)])
    else:
        #elimine la mauvaise porte restante
        porte=[bonne_porte]
    
    if strategie == 1:
        pass
    else:
        premier_choix = porte[0]
        
    return premier_choix==bonne_porte


def play_games(strategie,nb_tours):
    return [int(play_game(strategie)) for i in range(nb_tours)]


lch1 = play_games(1,3)
lch2 = play_games(2,1000)

print("le joueur a gagné {} fois sur 1000 en choisissant de garder son premier choix".format(sum(lch1)))
print("le joueur a gagné {} fois sur 1000 en choisissant de changer son premier choix".format(sum(lch2)))


l=play_games(2,10)
#plot = plt.plot(l)
l2=[random(),random(),random(),random(),random(),random()]
l3=[random(),random(),random(),random(),random(),random()]
print(l2)
plot1 = plt.plot(["a","b","c"],lch1)
```

```python
nbParties  = [100,1000,2000,5000,8000,10000]
resch1 = []
resch2 = []
for x in nbParties:
    resch1.append(play_games(1,x))
    resch2.append(play_games(2,x))

plt.figure()
plot = plt.scatter(nbParties,[sum(x) for x in resch1],color="green")
plot = plt.scatter(nbParties,[sum(x) for x in resch2],color="red")
```

```python
moych1 = []
moych2 = []
for i in range(6):
    moych1.append(sum(resch1[i])/nbParties[i])
    moych2.append(sum(resch2[i])/nbParties[i])

plot = plt.scatter(nbParties,moych1,color="green")
plot = plt.scatter(nbParties,moych2,color="red")
plt.figure()
plot = plt.bar([1,2],[sum(moych1),sum(moych2)],tick_label=["moyenne garder","moyenne changer"], color=["green","red"])
```
